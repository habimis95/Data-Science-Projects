```python
import pyspark
from pyspark import SparkContext
from pyspark.sql import SparkSession
```

```python
sc = SparkContext()
```

```python
#1. Read data => data. How many elements in data? Print first 5 element of the data
data = sc.textFile("Cung cap du lieu buoi 2/5000_points.txt", minPartitions = 3)
print("The type of data is", type(data))
```

```python
print("Number of elements:", data.count())
```

```python
data.take(5)
```

```python
#2. Create PairRDD name as pair_data from the data above, each element of the data will create 1
pair_data = data.map(lambda s: (int(s.split('\t')[0]), int(s.split('\t')[1])))
```

```python
pair_data.take(5)
```

```python
#3. Create pair_data_sort from pair_data in which key is sorted descending. Print first 5 element.
pair_data_sort = pair_data.sortByKey(ascending=False)
```

```python
pair_data_sort.take(5)
```

```python
#4. With pair_data_sort, lets count the number of items by key. Print out the key in which having number of item > 1
for key, val in pair_data_sort.countByKey().items():
    if val > 1:
        print(key, val)
```

```python
#5. Create pair_data_groupby from pair_data_sort by group value having same key. Print out the key in which having number of item > 1
pair_data_groupby = pair_data.groupByKey().collect()
```

```python
for x, y in pair_data_groupby:
    if len(list(y)) > 1:
        print(x, list(y))
```

```python
#6. Create pair_data_reduce from pair_data_sort in which value having same key, return max value from those values. How many elements of this pair_data_reduce ?
import math
pair_data_reduce = pair_data.reduceByKey(lambda x, y: max(x,y))
```

```python
print("Number of elements:", pair_data_reduce.count())
```

```python
# check 
for x, y in pair_data_reduce.collect():
    if x==620330:
        print(x, y)
```
