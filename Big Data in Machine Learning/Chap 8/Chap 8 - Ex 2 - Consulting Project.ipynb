```python
# Tree methods Example
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('dogfood').getOrCreate()
```

```python
# Load training data
data = spark.read.csv('Cung cap du lieu buoi 6/dog_food.csv', inferSchema=True, header=True)
```

```python
# Import VecrotAssembler and Vectors
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler
```

```python
data.columns
```

```python
assembler = VectorAssembler(inputCols=['A', 'B', 'C', 'D'],
                            outputCol='features')
```

```python
output = assembler.transform(data)
```

```python
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.classification import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier(labelCol='Spoiled', featuresCol='features')
```

```python
output.printSchema()
```

```python
final_data = output.select('features', 'Spoiled')
final_data.head()
```

```python
dtc_model = dtc.fit(final_data)
```

```python
dtc_model.featureImportances
```

Feature at index 2 (Chemical C) is by far the most important feature, meaning it is causing the early spoilage! This is a pretty interesting use of a machine learning model in an alternative way!
