# PIE CHART

```r
# Create data for the graph
x = c(46, 37, 35, 53)
labels = c('10A1','10A2','10A03','10A04')
# Give the chart file a name
# png(file = 'Chap 11/Hinh/classes_pie.png')
# plot the chart 
pie(x,labels)
# Save the file
dev.off()
```

```r
# Give the chart file a name
# png(file='Chap 11/Hinh/classes_pie_colors.jpg')
# Plot the chart with title and rainbow color pallet
# Length of the pallet should be same as the number of values => length(x)
pie(x, labels, main='Classes pie chart',
   col = rainbow(length(x)),
   clockwise = TRUE)
# Save the file
dev.off()
```

```r
# percentage
piepercent = round(100*x/sum(x),1)
# Give the chart file a name
# png(file='Chap 11/Hinh/classes_pie_percentage_legends.jpg')
# Plot the chart
pie(x, labels = paste(piepercent,'%'), main='Classes pie chart',
   col = rainbow(length(x)), clockwise = TRUE)
# Insert legends for chart
legend('topright', labels, cex = 0.8, fill = rainbow(length(x)))
# Save the file
dev.off()
```

```r
# percentage
piepercent = round(100*x/sum(x),1)
# Give the chart file a name
# png(file='Chap 11/Hinh/classes_pie_percentage_legends.jpg')
# Plot the chart
pie(x, labels = paste(piepercent,'%', '&',labels), main='Classes pie chart',
   col = rainbow(length(x)), clockwise = TRUE)
# Insert legends for chart
#legend('topright', labels, cex = 0.8, fill = rainbow(length(x)))
# Save the file
dev.off()
```

# 3D PIE CHART

```r
# 3D Pie Chart: to view the contribution of each factor
# Get the library.
library(plotrix)
# Give the chart file a name
# png(file = 'Chap 11/Hinh/classes_pie_3D.jpg')
# Plot the chart
pie3D(x, labels=labels, main = 'Classes Pie Chart',
     col = rainbow(length(x)))
# Save the file
dev.off()
```

# BAR CHART

```r
# Create the data for the chart 
H = c(5, 15, 20, 5)
M = c('Excellent','Very Good','Good','Medium')
# Give the chart file a name
# png(file = 'Chap 11/Hinh/10A1_grade_barchart.png')

# Plot the bar chart
barplot(H, names.arg=M, xlab='Grade', ylab='Students',
       col = 'blue', main='Grade of class 10A1', border='red')
# Save the file
dev.off()
```

```r
# Create the input vectors
colors = c('green','orange','brown','red') # for 4 grades
classes = c('10A1', '10A2', '10A3')
grades = c('Excellent','Very Good','Good','Medium')
# Create the matrix of the values
values = matrix(c(5,6,7,15,14,13,20,21,22,5,4,3),
               nrow=4, ncol=3, byrow=TRUE)
print(values)
# Give the chart file a name
# png(file = 'Chap 11/Hinh/grade_barchart_stacked.png')

# Create the bar chart
barplot(values, main = 'Grade of 3 classes', names.arg=classes, xlab='Class', ylab='Students',
       col = colors)
# Add the legend to the chart
legend('right', grades, cex=0.9, fill=colors)
# Save the file
dev.off()
```

# BOX PLOT

```r
data = read.csv('baseball.csv')
head(data)
```

```r
boxplot(data$height, main='Height', ylab="(m)")
head(data)
```

```r
boxplot(data$height,
        notch = TRUE,
        varwidth = TRUE,
        col = c('green'),
        main='Height', ylab="(m)")
```
