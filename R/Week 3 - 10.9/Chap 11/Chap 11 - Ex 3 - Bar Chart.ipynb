```r
year = c(2014, 2015, 2016, 2017)
first_fallen = c(309, 271, 263, 255)
one_hundred_fallen = c(309, 314, 312, 304)
df = data.frame(year,first_fallen, one_hundred_fallen)
```

```r
# group bar plot
barplot(t(df[c('first_fallen','one_hundred_fallen')]), beside = TRUE, 
        col = c("darkgreen","orange"), ylab = 'Number of leaf fall', xlab = 'Year',
        main = 'First and last dates of leaf fall in Yellow Birch',
        names.arg=df$year)
# Insert legends for chart
legend('bottomright', c('first_fallen','one_hundred_fallen'),cex=0.6, fill = c("darkgreen","orange"))
```
