```r


# Create the data for the chart.
year = c(2014, 2015, 2016, 2017)
first_fallen = c(309, 271, 263, 255)
one_hundred_fallen = c(309, 314, 312, 304)
df=data.frame(year, first_fallen, one_hundred_fallen)
# Give the chart file a name.
# png(file = "line_chart_2_lines.jpg")

# Plot the line chart.
plot(df$first_fallen,type = "o",col = "red",xlab='Year', ylab = "Number", ylim=c(250,320),
   main = "Leaves had fallen")
box()
lines(df$one_hundred_fallen, type = "o", col = "blue")
axis(1, at=1:4 ,labels = c('2014', '2015', '2016', '2017') )#as.character(df$year)
# legend
legend('topleft', c('first fallen leaf','one hundred fallen leaf'), cex=0.8, fill=c('red','blue'))
# Save the file.
dev.off()
```

```r


# Create the data for the chart.
year = c(2014, 2015, 2016, 2017)
first_fallen = c(309, 271, 263, 255)
one_hundred_fallen = c(309, 314, 312, 304)
df=data.frame(year, first_fallen, one_hundred_fallen)
# Give the chart file a name.
# png(file = "line_chart_2_lines.jpg")

# Plot the line chart.
plot(as.integer(df$year),df$first_fallen,type = "o",col = "red",xlab='Year', ylab = "Number", ylim=c(250,320), xlim=c(2014,2017),
   main = "Leaves had fallen")
box()
lines(as.integer(df$year),df$one_hundred_fallen, type = "o", col = "blue")
axis(1, at=1:4 ,labels = c('2014', '2015', '2016', '2017') )#as.character(df$year)
# legend
legend('topleft', c('first fallen leaf','one hundred fallen leaf'), cex=0.8, fill=c('red','blue'))
# Save the file.
dev.off()
```
