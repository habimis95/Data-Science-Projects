```r
library('plotrix')
```

```r
Tree_species = c('Beech', 'Chestnut','Hawthm','Red maple','Witch Hazel','Yellow Birch')
number_of_trees = c(2,2,1,4,1,1)
```

```r
# percentage
piepercent = round(100*number_of_trees/sum(number_of_trees),1)
#pie chart
pie3D(number_of_trees, labels=paste(piepercent,'%'), main = 'Distribution of tree species', col=rainbow(length(Tree_species)))
# insert legends for chart
legend('topright', Tree_species, cex=0.8, fill=rainbow(length(Tree_species)))
```

```r
barplot(number_of_trees, names.arg=Tree_species, xlab='Species', ylab = 'Number', col='darkgreen',
       main = 'Distribution of tree species', border='orange')
```
