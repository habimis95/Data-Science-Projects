```r
# scatter plot
plot(x = faithful$eruptions, y = faithful$waiting,
     xlab='Eruptions', ylab='Waiting', col=c("red","blue"),
     
     main='Correlation between eruptions & waiting')
# legend
legend(x="topright", c('Eruptions','waiting'), cex=0.8, fill=c("red","blue"))
# regression line
abline(lm(faithful$waiting ~ faithful$eruptions, data = faithful), col = "orange")
```
