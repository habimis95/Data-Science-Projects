```r
Temp = airquality$Temp
```

```r
hist(Temp,  main = 'Maximum daily temperature at La Guardla Airport', xlab = 'Temperature in degrees Fahrenheit', ylab='Frequency',
     labels = TRUE,
    col='orange',breaks=5)
```

```r
# Switch from Frequency to Density
hist(Temp, main = 'Maximum daily temperature at La Guardla Airport', xlab = 'Temperature in degrees Fahrenheit', ylab='Density',
     labels = TRUE,
    col='orange',breaks=5,freq=FALSE)
```

```r
# Histogram with 9 column
hist(Temp,  main = 'Maximum daily temperature at La Guardla Airport', xlab = 'Temperature in degrees Fahrenheit', ylab='Frequency',
     labels = TRUE,
    col='orange',breaks=9,xlim=c(50,100))
```

```r
hist(Temp,
main="Maximum daily temperature at La Guardia Airport",
xlab="Temperature in degrees Fahrenheit",
xlim=c(50,100),
col="orange",
border="brown",
breaks=c(55,60,70,75,80,100)
)
```
