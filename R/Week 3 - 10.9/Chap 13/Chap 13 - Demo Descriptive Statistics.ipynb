# Centrality measures
+ Mean

```r
data = read.csv('baseball.csv')
head(data)
```

```r
mean(data$height)
```

```r
library('ggplot2')
library('ggExtra')
library('gridExtra')
```

```r
p1 = ggplot(data, aes(x=data$height, fill=X, color=X)) +
        geom_density(alpha=0.1) + 
        theme_bw(base_size = 14)
p1
```

```r
p2 = ggplot(data, aes(x=data$weight, fill=X, color=X)) +
        geom_density(alpha=0.1) + 
        theme_bw(base_size = 14)
p2 
```

```r
library('GGally')
```

```r
grid.arrange(p1, p2, ncol=2)
```

- Mode

```r
# Create the function 
getmode = function(v){
    uniqv = unique(v)
    uniqv[which.max(tabulate(match(v, uniqv)))]
}
```

```r
getmode(data$height)
```

```r
getmode(data$weight)
```
