```r
library('ggplot2')
library('gridExtra')
library('ggExtra')
library('tidyverse')
```

```r
data2 = data %>% filter(Date==2001.25)
head(data2)
```

```r
scatter_p = ggplot(data = data2, aes(x  = Land.Value, y = Structure.Cost, fill = region, color = region, lwd = 1,label=State))
```

```r
scatter_p1 = scatter_p + geom_point() +
theme_bw(base_size=14)
scatter_p1
```

```r
scatter_p2 = scatter_p + geom_point() +
theme_bw(base_size=14)+
  stat_function(fun = dlnorm, size=1)+
  geom_text(aes(label=State),hjust=0, vjust=0)
scatter_p2 
```
