```r
height = seq(140, 180, 0.5)
print('Samples:')
print(height)
value = dnorm(height, 156, 4.6)
```

```r
plot(height, value, 
     type='l',
     ylab='f(height)',
     xlab='Height',
     main='Probability distribution of height in Vietnamese women')
```

```r
# => nu cao 1.6m thi xac suat se la
print('Probability of height = 160cm')
print(dnorm(160, mean=156, sd = 4.6))
```
