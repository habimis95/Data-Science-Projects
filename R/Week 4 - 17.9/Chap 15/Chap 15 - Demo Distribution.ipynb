# Normal Distribution
- dnorm(x, mean, sd): density

```r
x = seq(-10,10, by=.1)

# Choose the mean as 0 and standard deviation as 1
y = dnorm(x, mean = 0, sd = 1)

# Give the chart file a name
# png(file='')

plot(x,y)
```

- pnorm(x, mean, sd): probability

```r
x = seq(-10,10,by=.1)

y = pnorm(x, mean = 0, sd = 1)

plot(x, y)
```

- qnorm(x, mean, sd): quantile

```r
x = seq(0,1,by=0.02)

y = qnorm(x, mean = 0, sd = 1)

plot(x, y)
```

- rnorm(x, mean, sd): random

```r
y = rnorm(50, 165, sd=3)

hist(y)
```

# Binomial Distribution
- dbinom(x, size, prob)

```r
# create a sample of 50 numbers which are incremented by 1
x = seq(0, 50, by = 1)

# Create the binomial distribution
y = dbinom(x, 50, 0.5)

# Plot the graph
plot(x, y)
```

Example: There are 10 member in which 6 of them are women. If 3 are selected, what is the probability of select 2 women?

```r
p1 = dbinom(2, 3, 0.6)
p1
```

- pbinom(x, size, prob): return probability of an event, which is a binary value represent for probability (Cumulative distribution function or CDF)

```r
# probability of getting 25 or less heads from 50 tosses of a coin
p_x_25_less = pbinom(25, 50, 0.5)
p_x_25_less
```

```r
# probability of getting 25 or more heads from 50 tosses of a coin
p_x_25_more = 1 - pbinom(24, 50, 0.5)
p_x_25_more
```

- qbinom(x, size, prob): 

```r
# How many heads will have a probability of 0.25 will come out when a coin is tossed 50 times,
x = qbinom(0.25, 50, 0.5)
x
```

```r
x = rbinom(8, 100, 1/6)
x
```
