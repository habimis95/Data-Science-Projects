## Scatter plot - Regplot - Residual plot
- Cho dữ liệu salaries.xlsx.
- Biểu diễn mối quan hệ giữa thâm niên và lương

```python
data = pd.read_excel("salaries.xlsx")
```

```python
data.head()
```

```python
data.tail()
```

```python
# Xem xét mối quan hệ giữa số năm làm việc và lương
plt.figure(figsize=(8,6))
plt.scatter(data.years, data.salary)
plt.title("Relationship between Salary and Years", fontsize=18, color = 'b')
plt.xlabel("Years")
plt.ylabel("Salary")
plt.show()
```

```python
sns.lmplot(x="years", y="salary", data=data)
```

```python
plt.figure(figsize=(8,6))
sns.regplot(x="years", y="salary", data=data)
plt.title("Relationship between Salary and Years", fontsize=18, color = 'b')
```

#### Nhận xét:
- Dữ liệu có vẻ tuyến tính. Nhưng Phương sai ở những người >15 năm kinh nghiệm cao hơn <15 năm

```python
plt.figure(figsize=(8,6))
sns.residplot(x="years", y="salary", data=data)
```

## Pie chart - Tree map
- Cho dữ liệu là danh sách ứng viên và số phiếu bầu trong cuộc bầu cử tổng thống Mỹ năm 2016.
- Vẽ biểu đồ thể hiện tỷ lệ tổng số phiếu bầu lần lượt cho ứng viên

```python
df = pd.DataFrame(
    {
        'Name': ['Hillary Clinton', 'Donald Trump', 'Others'],
        'Virginia': [1981473, 1769443, 233715],
        'Maryland': [1677928, 943169, 160349],
        'West Virginia': [188794, 489371, 36258],
    }
)
```

```python
df
```

```python
df['Total'] = df['Virginia'] + df['Maryland'] + df['West Virginia']
df
```

```python
import squarify 
```

```python
plt.figure(figsize=(6,6))
plt.pie(df.Total, labels = df.Name, counterclock=False, shadow=True, autopct='%.0f%%')
plt.title("Presidential Election Results", fontsize=18, color = 'b')
plt.show()
```

```python
plt.figure(figsize=(8,6))
labels = ["%s\n(%.0f%%)" % (label) for label in zip(df.Name, df.Total/sum(df.Total)*100)]
squarify.plot(sizes=df.Total.values, label=labels, color=['orange', 'lightblue', 'green'])
plt.title("Presidential Election Results", fontsize=18, color = 'b')
plt.show()
```

```python
plt.figure(figsize=(8,6))
bar = sns.barplot(data=df, x="Name", y= "Total")
bar.set_xticklabels(bar.get_xticklabels(), rotation=90)
plt.title("Presidential Election Results", fontsize=18, color = 'b')
plt.show()
```

## Word Cloud
- Cho dữ liệu tweets_data_science.csv
- Biểu diễn thông tin của cột sentence lên biểu đồ

```python
ds = pd.read_csv("tweets_data_science.csv", index_col=0)
ds.head()
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
comment_words = ' '
stopwords = set(STOPWORDS) 
```

```python
for content in ds.sentence:       
    # convert content to string
    content = str(content) 
    comment_words = comment_words + ' ' + content
```

```python
wc = WordCloud(
    background_color='black',
    max_words=1000,
    stopwords=stopwords
)

# generate the word cloud
wc.generate(comment_words)
```

```python
# display the word clouds
plt.figure(figsize=(10, 12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
list_of_words = ['xe3', 'x81', 'RT', 'x80', "xa6'", 'xd9', 'co', 'xe2',
                'x80', "'", 'x82']
for word in list_of_words:
    stopwords.add(word) # add the less important word to stopwords 
```

```python
wc = WordCloud(
    background_color='black',
    max_words=300,
    stopwords=stopwords
)

# generate the word cloud
wc.generate(comment_words)
```

```python
# display the word clouds
plt.figure(figsize=(10, 12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
