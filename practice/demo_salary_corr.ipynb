```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_excel("salaries.xlsx")
```

```python
data.head()
```

```python
print("The variance of the Years:", data["years"].var())
print("The variance of the Salary:", data["salary"].var())
```

```python
sns.lmplot(x="years", y="salary", data=data)
```

# Nhận xét: 
- Có vẻ tuyến tính. Nhưng Phương sai ở những người >15 năm kinh nghiệm cao hơn <15 năm

```python
sns.distplot(data["years"])
```

```python
data["years"].skew() # lệch trái
```

```python
sns.distplot(data["salary"])
```

```python
data["salary"].skew() # lệch phải
```

```python
sns.residplot(x="years", y="salary", data=data)
```

# Correlation & kết luận

```python
# Pearson
data['years'].corr(data['salary']) 
```

```python
data['years'].corr(data['salary'], method='spearman') 
```

```python
from scipy import stats
```

```python
stats.pearsonr(data['years'], data['salary'])
```

```python
stats.spearmanr(data['years'], data['salary'])
```

#### Kết luận:
- Ho: Giả thuyết là years và salary không correlation
- Nhưng: p ~ 0 < 0.05 (alpha) => bác bỏ Ho
- => years và salary có correlation
